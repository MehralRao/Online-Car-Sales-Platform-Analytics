```python
df['Units Available'] = [random.randint(1, 5) for _ in range(len(df))]
```

```python
df
```

```python
df['Inventory Value'] = df['Price'] * df['Units Available']
```

```python
df
```

```python
print('Columns Explanation:\n')
print("1. 'Units Available': Represents the number of cars of a particular model that are currently in stock (inventory)")
print("2. 'Inventory Value': This shows the total value of the cars available for a specific model. It is calculated by multiplying the price of one car by the number of cars available.")
```

```python
print('Total Cars Availabe: ', df['Units Available'].sum())
```

```python
def mileage_category(mileage):
    if mileage < 30000:
        return 'Low'
    elif mileage < 60000:
        return 'Medium'
    else:
        return 'High'
df['Mileage Category'] = df['Mileage'].apply(mileage_category)
```

```python
df
```

```python
total_cars_per_brand = df.groupby('Brand')['Units Available'].sum().reset_index()
print(total_cars_per_brand)
```

```python
# Total Inventory Value and Units Available per Brand
brand_performance = df.groupby('Brand').agg({
    'Inventory Value': 'sum',  # Sum of Inventory Value for each brand
    'Units Available': 'sum'  # Sum of Units Available for each brand
}).reset_index()

# Total Inventory Value and Units Available per Model
model_performance = df.groupby('Model').agg({
    'Inventory Value': 'sum',  # Sum of Inventory Value for each model
    'Units Available': 'sum'  # Sum of Units Available for each model
}).reset_index()

print(brand_performance)
print('\n')
print(model_performance)
```

```python
def price_category(price):
    if price < 15000:
        return 'Low'
    elif price < 25000:
        return 'Medium'
    else:
        return 'High'
df['Price Category'] = df['Price'].apply(price_category)
```
